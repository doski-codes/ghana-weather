```python
# Import libraries
import numpy as np
import pandas as pd
from matplotlib.pyplot import plot as plt
```

```python
# Read data from station "3"
station3_data = pd.read_csv('data/3/TA00647.csv')
station3_metadata = pd.read_csv('data/3/metadata.csv')
```

```python
station3_data.rename(columns={'timestamp (UTC)': 'timestamp_(UTC)', 
                              'atmosphericpressure (kPa)': 'atmospheric_pressure_(kPa)', 
                              'precipitation (mm)': 'precipitation_(mm)',
                              'radiation (W/m2)': 'radiation_(W/m2)', 
                              'relativehumidity (-)': 'relative_humidity_(-)',
                              'temperature (degrees Celsius)': 'temperature_(degrees_Celsius)', 
                              'winddirection (degrees)': 'wind_direction_(degrees)',
                              'windgusts (m/s)': 'wind_gusts_(m/s)', 
                              'windspeed (m/s)': 'wind_speed_(m/s)'
                             }, inplace=True)
```

```python
station3_data.info()
```

```python
station3_data.duplicated().sum()
```

```python
station3_data['timestamp_(UTC)'] = pd.to_datetime(station3_data['timestamp_(UTC)'])
```

```python
station3_data.info()
```

```python
station3_data.describe()
```

```python
# Drop rows with any NA value to avoid us guessing values that might be wrong
station3_data.dropna(inplace=True)
```

```python
station3_metadata
```

```python
station3_metadata.columns
```

```python
station3_metadata.rename(columns={'station code': 'station_code', 
                                  'installation height (m)': 'installation_height_(m)',
                                  'elevation (m)': 'elevation_(m)'}, inplace=True)
```

```python
station3_code = station3_metadata['station_code'].unique()[0]
```

```python
station3_data['station_code'] = station3_code
```

```python
station3_data.head()
```

```python
weather_df = station3_data.merge(station3_metadata, how='left', on='station_code')
```
