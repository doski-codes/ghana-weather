```python
# Import libraries
import numpy as np
import pandas as pd
from matplotlib.pyplot import plot as plt
```

```python
# Read data from station "2"
station2_data = pd.read_csv('data/2/TA00098.csv')
station2_metadata = pd.read_csv('data/2/metadata.csv')
```

```python
station2_data.rename(columns={'timestamp (UTC)': 'timestamp_(UTC)', 
                              'atmosphericpressure (kPa)': 'atmospheric_pressure_(kPa)', 
                              'precipitation (mm)': 'precipitation_(mm)',
                              'radiation (W/m2)': 'radiation_(W/m2)', 
                              'relativehumidity (-)': 'relative_humidity_(-)',
                              'temperature (degrees Celsius)': 'temperature_(degrees_Celsius)', 
                              'winddirection (degrees)': 'wind_direction_(degrees)',
                              'windgusts (m/s)': 'wind_gusts_(m/s)', 
                              'windspeed (m/s)': 'wind_speed_(m/s)'
                             }, inplace=True)
```

```python
station2_data.info()
```

```python
station2_data.duplicated().sum()
```

```python
station2_data['timestamp_(UTC)'] = pd.to_datetime(station2_data['timestamp_(UTC)'])
```

```python
station2_data.info()
```

```python
station2_data.describe()
```

```python
# Drop rows with any NA value to avoid us guessing values that might be wrong
station2_data.dropna(inplace=True)
```

```python
station2_metadata
```

```python
station2_metadata.columns
```

```python
station2_metadata.rename(columns={'station code': 'station_code', 
                                  'installation height (m)': 'installation_height_(m)',
                                  'elevation (m)': 'elevation_(m)'}, inplace=True)
```

```python
station2_code = station2_metadata['station_code'].unique()[0]
```

```python
station2_data['station_code'] = station2_code
```

```python
station2_data.head()
```

```python
weather_df = station2_data.merge(station2_metadata, how='left', on='station_code')
```
